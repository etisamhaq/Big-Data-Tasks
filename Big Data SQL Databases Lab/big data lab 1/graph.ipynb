```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
versions = ['v1', 'v2', 'v3']
mysql_insertion = [4.32, 22.91, 222.86]
mysql_query = [0.0063, 0.19452, 0.77462]
mssql_insertion = [6.43, 64.09, 543.38]
mssql_query = [0.01962, 0.11978, 0.14506]

# Create figure and axes for insertion time
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot insertion time
ax1.plot(versions, mysql_insertion, marker='o', label='MySQL Insertion Time', color='blue')
ax1.plot(versions, mssql_insertion, marker='o', label='MSSQL Insertion Time', color='green')
ax1.set_title('Insertion Time Comparison')
ax1.set_xlabel('Versions')
ax1.set_ylabel('Time (seconds)')
ax1.legend()
ax1.grid(True)

# Plot query time
ax2.plot(versions, mysql_query, marker='o', label='MySQL Query Avg Time', color='blue')
ax2.plot(versions, mssql_query, marker='o', label='MSSQL Query Avg Time', color='green')
ax2.set_title('Query Average Time Comparison')
ax2.set_xlabel('Versions')
ax2.set_ylabel('Time (seconds)')
ax2.legend()
ax2.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```
